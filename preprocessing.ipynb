```python
import pandas as pd
```

```python
join_keys = ["GeoID", "Borough"]
```

```python
internet = pd.read_csv("./data/raw/Internet_Master_Plan__Adoption_and_Infrastructure_Data_by_Neighborhood.csv")
internet["GeoID"] = internet["Neighborhood Tabulation Area Code (NTA Code)"]
internet["Borough"] = internet["Borough Name"]
internet["neighborhood_name"] = internet["Neighborhood Tabulation Area Name (NTA NAME)"]
internet.drop(columns=["Neighborhood Tabulation Area Code (NTA Code)", "Borough Name"], inplace=True)
internet["Borough"] = internet.apply(lambda row: row.Borough if row.Borough != "Bronx" else "The Bronx", axis = 1)
internet_cols_to_keep = [
    "GeoID",
    "Borough",
    "neighborhood_name",
    "Total Population",
    "Population Density (per Sq. Mi.)",
    "Total Number of Households",
    "Home Broadband Adoption (Percentage of Households)",
    "Mobile Broadband Adoption (Percentage of Households)",
    "Mobile Dependent Households (Percentage of Households)",
    "Residential Broadband Choice Average by NTA",
    "Percentage of Blocks with ISP Reported Speed Availability of >=900 Mbps",
    "Public Wi-Fi in NYC Count",
    "NYC Internet Master Plan Open Access Infrastructure Cost Comparison Coefficient",
]
internet = internet[internet_cols_to_keep]
internet
```

```python
soc = pd.read_excel("./data/raw/ACS5yrNTA/soc_2016acs5yr_nta.xlsx", sheet_name="SocData")
soc_cols_to_keep = [
    "GeoID",
    "Borough",
    "EngOnly1P",
    "LgOEng1P",
    "NtvP",
    "Fb1P",
    "EA_BchDHP",
]
soc = soc[soc_cols_to_keep]
soc
```

```python
demo = pd.read_excel("./data/raw/ACS5yrNTA/demo_2016acs5yr_nta.xlsx", sheet_name="DemData")
racial_cols = {
    "WtNHP": "White",
    "BlNHP": "Black or African American",
    "AIANNHP": "American Indian and Alaska Native",
    "AsnNHP": "Asian",
    "NHPINHP": "Native Hawaiian and Other Pacific Islander",
    "OthNHP": "Some other race",
    "Rc2plNHP": "Two or more races",
}
demo_data_to_keep = [
    "GeoID",
    "Borough",
    "PopU181P",
    "Pop65pl1P",
] + list(racial_cols.keys())

demo = demo[demo_data_to_keep]
demo["plurality_racial_group"] = demo[racial_cols].idxmax(axis=1)
demo["plurality_racial_group"] = demo["plurality_racial_group"].replace(racial_cols)
demo.drop(columns=list(racial_cols.keys()), inplace=True)
demo
```

```python
econ = pd.read_excel("./data/raw/ACS5yrNTA/econ_2016acs5yr_nta.xlsx", sheet_name="EconData")
econ_cols_to_keep = [
    "GeoID",
    "Borough",
    "PerCapIncE",
    "FamBwPvP",
    "MgBSciArtP",
    "SrvcP",
    "SalesOffP",
    "NRCnstMntP",
    "PrdTrnsMMP",
    "LFP",
]
econ = econ[econ_cols_to_keep]
econ["wfh_capable"] = econ["MgBSciArtP"] + econ["SalesOffP"]
econ["wfh_incapable"] = 100 - econ["wfh_capable"]

occupation_cols_to_drop = {
    'MgBSciArtP',
    'SrvcP',
    'SalesOffP',
    'NRCnstMntP',
    'PrdTrnsMMP',
}
econ.drop(columns=list(occupation_cols_to_drop), inplace=True)

econ
```

```python
housing = pd.read_excel("./data/raw/ACS5yrNTA/hous_2016acs5yr_nta.xlsx", sheet_name="HousData")
housing_cols_to_keep = [
    "GeoID",
    "Borough",
    "MdGRE",
    "MdVlE",
    "OOcHU1P",
    "ROcHUP",
    "OcHU1P",
]
housing = housing[housing_cols_to_keep]
housing
```

```python
data = internet
for right in [soc, demo, housing, econ]:
    data = data.merge(right, on=["GeoID", "Borough"], how="left")
data
```

```python
extra_column_mapping = {
    'EngOnly1P': 'English only language %',
    'LgOEng1P': 'Language other than English %',
    'NtvP': 'Native born in US %',
    'Fb1P': 'Foreign born %',
    'EA_BchDHP': 'Completed Bachelors degree or higher %',
    'PopU181P': 'Population under 18 years %',
    'Pop65pl1P': 'Population over 64 years %',
    'MdGRE': 'Median House gross rent (in $)',
    'MdVlE': 'Median House value (in $)',
    'OOcHU1P': 'Owner occupied Housing unit %',
    'ROcHUP': 'Renter occupied Housing unit %',
    'OcHU1P': 'Total occupied Housing unit %',
    'PerCapIncE': 'Per Capita Income (in $)',
    'FamBwPvP': 'Families below poverty level %',
    'LFP': 'Population employed in labor force %',
    
    # Renaming few to follow convention
    'Home Broadband Adoption (Percentage of Households)' :'Home Broadband Adoption %',
    'Mobile Broadband Adoption (Percentage of Households)' : 'Mobile Broadband Adoption %',
    'neighborhood_name' : 'Neighborhood Name',
    'Percentage of Blocks with ISP Reported Speed Availability of >=900 Mbps': 'Blocks with ISP Reported Speed Availability of >=900 Mbps in %',
    'wfh_capable': 'Work from Home - Capable',
    'wfh_incapable':'Work from Home - Incapable',
    
}
data.rename(columns=extra_column_mapping, inplace=True)
```

```python
irregular_neighborhoods = [
    "BK99", # park-cemetary-etc-Brooklyn
    "BX99", # park-cemetery-etc-Bronx
    "MN99", # park-cemetery-etc-Manhattan
    "QN99", # park-cemetery-etc-Queens
    "SI99", # park-cemetery-etc-Staten Island
    "BX98", # Rikers Island
    "QN98", # Airport
]
data_no_weird = data[~data["GeoID"].isin(irregular_neighborhoods)]
len(data_no_weird)
```

```python
data_no_weird.to_csv("./data/clean/combined_dataset_no_weird.csv", index=False)
data.to_csv("./data/clean/combined_dataset.csv", index=False)
```
